```python
# Import necessary libraries
import os
import json
import pandas as pd
import logging
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Flatten, Dropout, BatchNormalization, Input, Concatenate
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from transformers import DistilBertTokenizerFast, TFDistilBertModel
import numpy as np
import psutil

# Set up logger
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger()

# Paths for Annotations and Questions
train_annotations_path = r"data/DeepLearning/Annotations/Annotations_Train_abstract_v002/abstract_v002_train2015_annotations.json"
train_image_folder = r'data/DeepLearning/Images/scene_img_abstract_v002_train2015'  # Correct folder path
val_image_folder = r'data/DeepLearning/Images/scene_img_abstract_v002_val2015'  # Correct folder path

# Paths for Questions (Train, Val)
train_questions_open_ended_path = r"data/DeepLearning/Questions/Questions_Train_abstract_v002/OpenEnded_abstract_v002_train2015_questions.json"
train_questions_multiple_choice_path = r"data/DeepLearning/Questions/Questions_Train_abstract_v002/MultipleChoice_abstract_v002_train2015_questions.json"
val_questions_open_ended_path = r"data/DeepLearning/Questions/Questions_Val_abstract_v002/OpenEnded_abstract_v002_val2015_questions.json"

# Function to load JSON data
def load_json_file(filepath):
    try:
        with open(filepath, 'r', encoding='utf-8') as f:
            data = json.load(f)
            logger.info(f"File loaded successfully: {filepath} with keys: {list(data.keys())}")
            return data
    except FileNotFoundError:
        logger.error(f"File not found: {filepath}")
        raise
    except json.JSONDecodeError:
        logger.error(f"Error decoding JSON from file: {filepath}")
        raise

# Load JSON data
logger.info("Loading training, validation, and testing questions and annotations.")
train_questions_open_ended = load_json_file(train_questions_open_ended_path)
train_questions_multiple_choice = load_json_file(train_questions_multiple_choice_path)
train_annotations = load_json_file(train_annotations_path)
val_questions_open_ended = load_json_file(val_questions_open_ended_path)

# Convert JSON data into DataFrames
df_train_questions_open_ended = pd.DataFrame(train_questions_open_ended['questions'])
df_train_annotations = pd.DataFrame(train_annotations['annotations'])
df_val_questions_open_ended = pd.DataFrame(val_questions_open_ended['questions'])

# Create DataFrame for Multiple Choice
df_train_questions_multiple_choice = pd.DataFrame(train_questions_multiple_choice['questions'])

# Update the 'file_name' to include the correct folder structure
df_train_questions_multiple_choice['file_name'] = df_train_questions_multiple_choice['image_id'].apply(
    lambda x: os.path.join(train_image_folder, f'abstract_v002_train2015_{{:012d}}'.format(int(x)) + '.png')
)

# Plot 1: Distribution of questions across image IDs
plt.figure(figsize=(10, 6))
sns.countplot(data=df_train_questions_multiple_choice, x='image_id', order=df_train_questions_multiple_choice['image_id'].value_counts().index[:20])
plt.xticks(rotation=90)
plt.title('Top 20 Images with Most Questions')
plt.xlabel('Image ID')
plt.ylabel('Number of Questions')
plt.show()

# Plot 2: Distribution of question types (based on question length for example)
df_train_questions_multiple_choice['question_length'] = df_train_questions_multiple_choice['question'].apply(len)

plt.figure(figsize=(10, 6))
sns.histplot(df_train_questions_multiple_choice['question_length'], bins=30, kde=True)
plt.title('Distribution of Question Lengths')
plt.xlabel('Length of Questions')
plt.ylabel('Frequency')
plt.show()

# Plot 3: Relationship between number of questions and image usage
image_question_counts = df_train_questions_multiple_choice['image_id'].value_counts()

plt.figure(figsize=(10, 6))
sns.histplot(image_question_counts, bins=30, kde=True)
plt.title('Distribution of Number of Questions per Image')
plt.xlabel('Number of Questions per Image')
plt.ylabel('Frequency')
plt.show()

# Check the structure of the data
print(df_train_questions_multiple_choice.head())

# Sampling 2% of the data
df_subset = df_train_questions_multiple_choice.sample(frac=0.02, random_state=42)

# Log information on the selected subset
logger.info(f"Selected {len(df_subset)} questions for training (2% of the dataset).")

# Data Cleaning and Checking
def check_missing_values(df, df_name):
    missing_values = df.isnull().sum()
    logger.info(f"Missing values in {df_name}: {missing_values}")

# Check for missing values
check_missing_values(df_train_questions_multiple_choice, 'Training Multiple-Choice Questions')
check_missing_values(df_train_annotations, 'Training Annotations')

logger.info("Data loading and cleaning completed successfully.")

# Ensure the 'file_name' column is created properly
def create_file_name(df, image_prefix, file_format='.png'):
    df['file_name'] = df['image_id'].apply(lambda x: f'{image_prefix}_{{:012d}}'.format(int(x)) + file_format)
    return df

# Ensure the file names for training and validation sets
df_train_questions_open_ended['file_name'] = df_train_questions_open_ended['image_id'].apply(lambda x: os.path.join(train_image_folder, f'abstract_v002_train2015_{{:012d}}'.format(int(x)) + '.png'))
df_val_questions_open_ended['file_name'] = df_val_questions_open_ended['image_id'].apply(lambda x: os.path.join(val_image_folder, f'abstract_v002_val2015_{{:012d}}'.format(int(x)) + '.png'))

# Check if the file names are correctly generated
print(df_train_questions_open_ended[['image_id', 'file_name']].head())
print(df_val_questions_open_ended[['image_id', 'file_name']].head())

# Tokenizer and Pre-trained DistilBERT Model
logging.info("Loading DistilBERT tokenizer and model...")
tokenizer = DistilBertTokenizerFast.from_pretrained('distilbert-base-uncased')
distilbert_model = TFDistilBertModel.from_pretrained('distilbert-base-uncased', trainable=False)
logging.info("DistilBERT model and tokenizer loaded.")

# Function to tokenize questions with dynamic truncation and logging
def tokenize_questions(questions, max_length=128):
    try:
        tokenized = tokenizer(questions, padding='max_length', truncation=True, max_length=max_length, return_tensors='tf')
        return tokenized['input_ids'], tokenized['attention_mask']
    except Exception as e:
        logging.error(f"Error tokenizing questions: {e}")
        return np.zeros((len(questions), max_length)), np.zeros((len(questions), max_length))

# EfficientNetB0 for image encoding
logging.info("Loading EfficientNet-B0 model for image encoding...")
image_model = EfficientNetB0(include_top=False, input_shape=(224, 224, 3), pooling='avg')
logging.info("EfficientNet-B0 model loaded.")

# Custom layer for DistilBERT
class DistilBertLayer(tf.keras.layers.Layer):
    def __init__(self, distilbert_model, **kwargs):
        super(DistilBertLayer, self).__init__(**kwargs)
        self.distilbert_model = distilbert_model

    def call(self, inputs):
        text_input_tensor, attention_input_tensor = inputs
        output = self.distilbert_model(input_ids=text_input_tensor, attention_mask=attention_input_tensor).last_hidden_state[:, 0, :]
        return output

# Inputs
image_input = Input(shape=(224, 224, 3), name='image_input')  # Image input
text_input = Input(shape=(None,), dtype=tf.int32, name='text_input')  # Tokenized text input (IDs)
attention_input = Input(shape=(None,), dtype=tf.int32, name='attention_input')  # Attention mask input

# DistilBERT for text encoding
text_embeddings = DistilBertLayer(distilbert_model)([text_input, attention_input])

# Combine image and text embeddings
combined_input = Concatenate()([tf.keras.layers.BatchNormalization()(image_model(image_input)), text_embeddings])

# Add dense layers for classification with dynamic number of classes
num_classes = 3000  # This can be parameterized based on dataset
x = Dense(64, activation='relu')(combined_input)
x = BatchNormalization()(x)
x = Dropout(0.5)(x)
output = Dense(num_classes, activation='softmax')(x)

# Create the final model
model = Model(inputs=[image_input, text_input, attention_input], outputs=output)

# Compile the model
model.compile(optimizer=Adam(learning_rate=1e-4), loss='categorical_crossentropy', metrics=['accuracy'])

# Model Summary
model.summary()

# Early stopping, learning rate reduction, and model checkpoint callbacks
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-7)
checkpoint = ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True)

# Updated generator function to ensure consistent batch size and better memory optimization
def generator(data_source, batch_size=1):
    logging.info("Starting data generation...")
    while True:  # Creating an infinite loop to ensure continuous data generation
        for i in range(0, len(data_source), batch_size):
            logger.info(f"Generating batch from index {i} to {i + batch_size}")  # Debugging statement to track batch index
            batch_data = data_source[i:i+batch_size]

            images = []
            question_tokens_list = []
            attention_mask_list = []
            labels = []

            for img_path, question, attention_mask, label in batch_data:
                abs_img_path = img_path
                try:
                    # Load the image
                    img = plt.imread(abs_img_path)[..., :3]  # Discard alpha channel if present, keep RGB channels
                    img = tf.image.resize(img, (224, 224))  # Resize image to match input shape
                    img = tf.cast(img, tf.float32) / 255.0  # Normalize the image
                except FileNotFoundError:
                    logging.warning(f"File not found: {abs_img_path}, using a placeholder image.")
                    img = np.zeros((224, 224, 3))  # Use a placeholder for missing files
                    img = tf.cast(img, tf.float32) / 255.0

                # Tokenize questions
                tokenized_question, tokenized_attention_mask = tokenize_questions([question], max_length=128)

                # Ensure the question tokens and attention masks are of type int32
                question_tokens = tf.cast(tokenized_question, tf.int32)
                attention_mask = tf.cast(tokenized_attention_mask, tf.int32)

                # Append to batch lists
                images.append(img)
                question_tokens_list.append(tf.squeeze(question_tokens))
                attention_mask_list.append(tf.squeeze(attention_mask))
                # Ensure labels are one-hot encoded for categorical crossentropy
                label_one_hot = tf.one_hot(label, depth=num_classes)
                labels.append(label_one_hot)

            # Convert lists to tensors
            images = tf.convert_to_tensor(images, dtype=tf.float32)
            question_tokens_list = tf.convert_to_tensor(question_tokens_list, dtype=tf.int32)
            attention_mask_list = tf.convert_to_tensor(attention_mask_list, dtype=tf.int32)
            labels = tf.convert_to_tensor(labels, dtype=tf.float32)

            # Print debug information about the shapes of the tensors
            logger.info(f"Batch shapes - images: {images.shape}, question_tokens: {question_tokens_list.shape}, attention_masks: {attention_mask_list.shape}, labels: {labels.shape}")

            # Print memory usage info
            logger.info(f"Memory usage: {psutil.virtual_memory().percent}%")

            # Yield the batch
            yield (images, question_tokens_list, attention_mask_list), labels

# Define the output signature to match the expected batch structure
def generator_signature(num_classes, batch_size=1):
    return (
        (
            tf.TensorSpec(shape=(None, 224, 224, 3), dtype=tf.float32),  # Batch of image tensors
            tf.TensorSpec(shape=(None, 128), dtype=tf.int32),  # Batch of tokenized question IDs
            tf.TensorSpec(shape=(None, 128), dtype=tf.int32)  # Batch of attention masks
        ),
        tf.TensorSpec(shape=(None, num_classes), dtype=tf.float32)  # Batch of one-hot encoded labels
    )

# Adjust batch size for better memory optimization
batch_size = 4  # Reduced to lower memory usage

# Use the subset data as the training data source
data_source = df_subset[['file_name', 'question', 'question', 'image_id']].values.tolist()

# Create the dataset using the generator and the actual data source
train_dataset = tf.data.Dataset.from_generator(
    lambda: generator(data_source, batch_size=batch_size),
    output_signature=generator_signature(num_classes)
)
train_dataset = train_dataset.prefetch(buffer_size=1)

# Create validation dataset using a portion of the data
df_val_subset = df_val_questions_open_ended.sample(frac=0.01, random_state=42)
val_data_source = df_val_subset[['file_name', 'question', 'question', 'image_id']].values.tolist()

val_dataset = tf.data.Dataset.from_generator(
    lambda: generator(val_data_source, batch_size=batch_size),
    output_signature=generator_signature(num_classes)
)
val_dataset = val_dataset.prefetch(buffer_size=1)  # Optimize data loading with prefetching

# Training the model with the updated generator
history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=5,
    callbacks=[early_stopping, reduce_lr, checkpoint],
    steps_per_epoch=max(len(data_source) // batch_size, 1)  # Ensure correct steps per epoch
)

# Plot training history
def plot_training_history(history):
    acc = history.history['accuracy']
    val_acc = history.history.get('val_accuracy', [])
    loss = history.history['loss']
    val_loss = history.history.get('val_loss', [])
    epochs = range(1, len(acc) + 1)

    plt.figure(figsize=(12, 6))

    # Plot accuracy
    plt.subplot(1, 2, 1)
    plt.plot(epochs, acc, 'bo-', label='Training accuracy')
    if val_acc:
        plt.plot(epochs, val_acc, 'ro-', label='Validation accuracy')
    plt.title('Training and validation accuracy')
    plt.legend()

    # Plot loss
    plt.subplot(1, 2, 2)
    plt.plot(epochs, loss, 'bo-', label='Training loss')
    if val_loss:
        plt.plot(epochs, val_loss, 'ro-', label='Validation loss')
    plt.title('Training and validation loss')
    plt.legend()

    plt.show()

# Plot the training history
plot_training_history(history)
```
